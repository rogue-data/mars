```python
import pandas as pd
```

```python
d12_15=pd.read_csv('d12_15.csv')
d15_18=pd.read_csv('d15_18.csv')
d18_21=pd.read_csv('d18_21.csv')
d21_24=pd.read_csv('d21_24.csv')
d24_27=pd.read_csv('d24_27.csv')
```

```python
data=[d12_15,d15_18,d18_21,d21_24,d24_27]
song_data=pd.concat(data)
song_data=song_data.drop('Unnamed: 0',axis=1)
song_data
```

```python
song_data.drop_duplicates(subset=['title'])
```

```python
drem=pd.read_csv('drem.csv')
```

```python
drem=drem.drop('Unnamed: 0',axis=1)
drem
```

```python
songs_dataset=[drem,song_data]
songs12_27k=pd.concat(songs_dataset)
```

```python
songs12_27k
```

```python
songs12_27k=songs12_27k.drop_duplicates(subset=['title'])
songs12_27k
```

```python
songs12_27k.to_csv('songs12_27k.csv')
```
